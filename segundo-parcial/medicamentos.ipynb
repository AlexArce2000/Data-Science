#1. Plantear la hipotesis

Definimos las hipótesis para este estudio:

Hipótesis Nula ($H_0$): No hay diferencia en la reducción de la presión arterial sistólica (PAS) entre el grupo de tratamiento (medicamento) y el grupo de control (placebo). Matemáticamente, esto lo expresaremos como:

$H_0:μ_{medicamento}=μ_{placebo}$

Hipótesis Alternativa ($H_1$): Hay una diferencia en la reducción de la PAS entre el grupo de tratamiento y el grupo de control. Esto lo expresaremos como:

$H_1:μ_{medicamento}≠μ_{placebo}$

Dado que queremos saber si el medicamento tiene algún efecto, sin especificar si aumentará o disminuirá la PAS en comparación con el placebo, este es un caso de prueba de dos colas.



#2. Seleccionar la prueba estadística apropiada

Debido a que queremos comparar las medias de dos muestras independientes (grupo de tratamiento y grupo de control) y no conocemos las desviaciones estándar de la población, una prueba t para muestras independientes es adecuada para este caso. La función ttest_ind de scipy.stats es ideal para realizar este análisis.

#3. Calcular el estadístico de prueba y el valor p

```python
import numpy as np
from scipy.stats import ttest_ind

# Datos de los dos grupos dados
grupo_medicamento = np.array([130, 125, 140, 118, 124, 134, 128, 136, 130, 129])
grupo_placebo = np.array([140, 142, 138, 150, 139, 142, 137, 148, 142, 141])

statistic, p_value = ttest_ind(grupo_medicamento, grupo_placebo)
print("Estadístico t:", statistic)
print("Valor p:", p_value)
```

Comparación del valor p con el nivel de significancia:
